```python
train_df = pd.read_csv("smart-city-traffic-patterns/train_aWnotuB.csv")
test_df = pd.read_csv("smart-city-traffic-patterns/datasets_8494_11879_test_BdBKkAj.csv")

# Combine train and test for uniform feature engineering
data = pd.concat([train_df.drop(columns=['Vehicles', 'ID']), test_df.drop(columns=['ID'])], ignore_index=True)
data['DateTime'] = pd.to_datetime(data['DateTime'])
data = data.set_index('DateTime')
```

```python
# --- 1. Basic Time-Series Features ---
data['Hour'] = data.index.hour
data['DayOfWeek'] = data.index.dayofweek
data['Month'] = data.index.month
data['Year'] = data.index.year

# Convert Junction to categorical/object type
data['Junction'] = data['Junction'].astype('category')
```

```python
# --- 2. Holiday Feature Engineering (Manual Approach) ---
# List of key Indian public holidays for the data period (2015-2017)
manual_holidays = [
    # 2015
    '2015-01-26', '2015-03-06', '2015-08-15', '2015-10-02', '2015-11-11', '2015-12-25',
    # 2016
    '2016-01-26', '2016-03-23', '2016-08-15', '2016-10-02', '2016-10-30', '2016-12-25',
    # 2017
    '2017-01-26', '2017-03-13', '2017-08-15', '2017-10-02', '2017-10-19', '2017-12-25',
]
holiday_dates = {pd.to_datetime(date).date() for date in manual_holidays}

# Create the 'IsHoliday' feature (1 if holiday, 0 otherwise)
data['IsHoliday'] = data.index.to_series().apply(lambda x: 1 if x.date() in holiday_dates else 0)
```

```python
# --- 3. Create a DayType feature (Weekday, Weekend, Holiday) ---
data['DayType'] = data['DayOfWeek']
data['DayType'] = data['DayType'].replace([5, 6], 2) # 2 for Weekend (Sat=5, Sun=6)
data['DayType'] = data['DayType'].apply(lambda x: 1 if x < 5 else x) # 1 for Weekday (Mon-Fri)
data.loc[data['IsHoliday'] == 1, 'DayType'] = 3 # 3 for Holiday (highest priority)

# Remove the original DayOfWeek column to avoid multicollinearity
data = data.drop(columns=['DayOfWeek'])

print("\n--- Features After Engineering ---")
print(data.head())
print(data['DayType'].value_counts())
```

```python
train_labels = train_df['Vehicles']
test_ids = test_df['ID']

# Extract features
data['Hour'] = data.index.hour
data['Month'] = data.index.month
data['Year'] = data.index.year
```

```python
indian_holidays = holidays.India(years=data.index.year.unique())
data['IsHoliday'] = data.index.to_series().apply(lambda x: 1 if x.date() in indian_holidays else 0)

data['DayOfWeek'] = data.index.dayofweek
data['DayType'] = data['DayOfWeek']
data['DayType'] = data['DayType'].replace([5, 6], 2) # 2 for Weekend
data['DayType'] = data['DayType'].apply(lambda x: 1 if x < 5 else x) # 1 for Weekday (Mon-Fri)
data.loc[data['IsHoliday'] == 1, 'DayType'] = 3 # 3 for Holiday
data = data.drop(columns=['DayOfWeek'])
```

```python
# Convert integer features to category for one-hot encoding
data['Junction'] = data['Junction'].astype('category')
data['Month'] = data['Month'].astype('category')
data['DayType'] = data['DayType'].astype('category')
```

```python
# One-Hot Encode categorical features
data_encoded = pd.get_dummies(data, columns=['Junction', 'Month', 'DayType'], drop_first=True)

# Split the data back into train and test sets
X_train = data_encoded.iloc[:len(train_labels)]
X_test = data_encoded.iloc[len(train_labels):]
y_train = train_labels

print("\n--- Encoded Training Data Head ---")
print(X_train.head())
print(f"\nTraining set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")
```

# Model Building

```python
# Initialize the XGBoost Regressor
# A good starting point for hyperparameters:
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror', # Standard objective for regression
    n_estimators=500,             # Number of boosting rounds (trees)
    max_depth=7,                  # Maximum tree depth
    learning_rate=0.05,           # Step size shrinkage
    subsample=0.8,                # Subsample ratio of the training instance
    colsample_bytree=0.8,         # Subsample ratio of columns when constructing each tree
    random_state=42,
    n_jobs=-1
)

# Train the model
print("\n--- Starting XGBoost Model Training ---")
xgb_model.fit(X_train, y_train, verbose=False)
print("XGBoost Model Training Complete.")
```

```python
# 1. Save the Trained XGBoost Model
model_filename = 'xgb_traffic_model.joblib'
joblib.dump(xgb_model, model_filename)

feature_cols = X_train.columns.tolist()
features_filename = 'model_features.joblib'
joblib.dump(feature_cols, features_filename)

print(f"\n Model saved as: {model_filename}")
print(f" Features list saved as: {features_filename}")
```

```python
# Get feature importances
feature_importances = pd.Series(xgb_model.feature_importances_, index=X_train.columns)
top_10_features = feature_importances.nlargest(10)

print("\n--- Top 10 Feature Importances ---")
print(top_10_features)
```

```python
# Generate predictions on the test set
test_predictions = xgb_model.predict(X_test)

# Ensure predictions are non-negative (as vehicle count cannot be negative)
test_predictions = np.maximum(0, test_predictions)

# Create the final submission DataFrame
submission_df = pd.DataFrame({
    'ID': test_ids,
    'Vehicles': test_predictions.round().astype(int) # Round and convert to integer count
})

# Display the first few rows of the submission file
print("\n--- Generated Submission File Head ---")
print(submission_df.head())

# Save the submission file to a CSV
submission_filename = 'traffic_prediction_xgboost_submission.csv'
submission_df.to_csv(submission_filename, index=False)

print(f"\n Project Complete! Your submission file is ready: **{submission_filename}**")
```

# Plotting the Visualizations

```python
train_df['DateTime'] = pd.to_datetime(train_df['DateTime'])
train_df = train_df.set_index('DateTime')


train_df['Hour'] = train_df.index.hour
train_df['DayOfWeek'] = train_df.index.dayofweek

# Manual Indian Holiday Definition (to avoid 'holidays' library error)
manual_holidays = [
    '2015-01-26', '2015-03-06', '2015-08-15', '2015-10-02', '2015-11-11', '2015-12-25',
    '2016-01-26', '2016-03-23', '2016-08-15', '2016-10-02', '2016-10-30', '2016-12-25',
    '2017-01-26', '2017-03-13', '2017-08-15', '2017-10-02', '2017-10-19', '2017-12-25',
]
holiday_dates = {pd.to_datetime(date).date() for date in manual_holidays}
train_df['IsHoliday'] = train_df.index.to_series().apply(lambda x: 1 if x.date() in holiday_dates else 0)

# Create 'DayType' feature (1:Weekday, 2:Weekend, 3:Holiday)
train_df['DayType'] = train_df['DayOfWeek'].replace({5: 2, 6: 2})
train_df.loc[train_df['DayType'] < 5, 'DayType'] = 1
train_df.loc[train_df['IsHoliday'] == 1, 'DayType'] = 3
```

```python
# Aggregation for Plot 1: Daily Pattern
hourly_traffic = train_df.groupby(['Junction', 'Hour'])['Vehicles'].mean().reset_index()

# Aggregation for Plot 2: Day Type Pattern
daytype_map = {1: 'Weekday', 2: 'Weekend', 3: 'Holiday'}
train_df['DayType_Label'] = train_df['DayType'].map(daytype_map)
daytype_traffic = train_df.groupby(['Junction', 'DayType_Label'])['Vehicles'].mean().reset_index()

# Pivot for grouped bar chart
daytype_traffic_pivot = daytype_traffic.pivot(index='DayType_Label', columns='Junction', values='Vehicles').fillna(0)
```

```python
junction_colors = {1: 'blue', 2: 'orange', 3: 'green', 4: 'red'}
day_order = ['Weekday', 'Weekend', 'Holiday']
junctions = sorted(train_df['Junction'].unique())
```

```python
# Plot 1: Average Hourly Traffic Volume (Daily Seasonality)
plt.figure(figsize=(12, 6))
for junction in junctions:
    subset = hourly_traffic[hourly_traffic['Junction'] == junction]
    plt.plot(subset['Hour'], subset['Vehicles'], label=f'Junction {junction}', 
            marker='o', linestyle='-', color=junction_colors.get(junction))

plt.title('Average Hourly Traffic Volume by Junction (Daily Pattern)', fontsize=14)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Average Vehicle Count', fontsize=12)
plt.xticks(range(24))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Junction', loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Day Type Traffic Pattern (Weekly and Holiday Seasonality)
plt.figure(figsize=(10, 6))
bar_width = 0.2
x = np.arange(len(day_order))

# Re-index to ensure correct order
daytype_traffic_pivot = daytype_traffic_pivot.reindex(day_order)

for i, junction in enumerate(junctions):
    heights = daytype_traffic_pivot[junction].values
    plt.bar(x + i * bar_width, heights, bar_width, 
           label=f'Junction {junction}', color=junction_colors.get(junction))

plt.title('Average Traffic Volume by Day Type (Weekday, Weekend, Holiday)', fontsize=14)
plt.xlabel('Day Type', fontsize=12)
plt.ylabel('Average Vehicle Count', fontsize=12)
plt.xticks(x + bar_width * (len(junctions) - 1) / 2, day_order)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Junction', loc='upper right')
plt.tight_layout()
plt.show()
```
